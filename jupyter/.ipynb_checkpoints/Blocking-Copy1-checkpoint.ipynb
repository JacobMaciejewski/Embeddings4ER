```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from numpy import ones
import scipy


import seaborn as sns
from utils import box_plot, filter_df, vectorizers_order, line_plot, heatmap_plot
```

```python
f_sch = ('Columns', 'ne', 2)
f_agn = ('Columns', 'eq', 2)
f_q2i = ('Direction', 'eq', 'q2i')
f_i2q = ('Direction', 'eq', 'i2q')
f_exa = ('Exact', 'eq', 'exact')
f_app = ('Exact', 'eq', 'approx')
f_k = ('k', 'eq', 10)
```

# Real

```python
results = pd.read_csv('../logs/blocking_recall_euclidean_real.csv')
results
```

```python
tuples = [('D1', 0), ('D2', 0), ('D3', 0), ('D4', 0), ('D5', 1), ('D6', 0), ('D7', 0), ('D8', 0), ('D9', 0), ('D10', 0), 
          ('D1', 2), ('D2', 2), ('D3', 2), ('D4', 2), ('D5', 2), ('D6', 2), ('D7', 2), ('D8', 2), ('D9', 2), ('D10', 2)]
case_order = [f'D{i}' for i in range(1, 11)]
results2 = results.loc[results.apply(lambda x: (x['Case'], x['Columns']) in tuples, axis=1)]
print(results2.shape)
results2 = filter_df(results2, [f_i2q, f_k])
print(results2.shape)
```

## How each case performs

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
df2 = filter_df(results2, [f_sch, f_exa])
box_plot(df2, 'Case', 'Recall', 'Vectorizer', angle_x=90, ax=axes, title='i2q Schema-Based Recall', order=case_order)
plt.savefig(f'../plots/blocking_real_exact_case_sch.pdf', bbox_inches='tight')

fig, axes = plt.subplots(nrows=1, ncols=1)
df2 = filter_df(results2, [f_sch, f_exa])
box_plot(df2, 'Case', 'Precision', 'Vectorizer', angle_x=90, ax=axes, title='i2q Schema-Based Precision', order=case_order)
plt.savefig(f'../plots/blocking_real_exact_case_sch2.pdf', bbox_inches='tight')

fig, axes = plt.subplots(nrows=1, ncols=1)
df2 = filter_df(results2, [f_agn, f_exa])
box_plot(df2, 'Case', 'Recall', 'Vectorizer', angle_x=90, ax=axes, title='i2q Schema-Agnostic Recall', order=case_order)
plt.savefig(f'../plots/blocking_real_exact_case_agn.pdf', bbox_inches='tight')

fig, axes = plt.subplots(nrows=1, ncols=1)
df2 = filter_df(results2, [f_agn, f_exa])
box_plot(df2, 'Case', 'Precision', 'Vectorizer', angle_x=90, ax=axes, title='i2q Schema-Agnostic Precision', order=case_order)
plt.savefig(f'../plots/blocking_real_exact_case_agn2.pdf', bbox_inches='tight')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
df2 = filter_df(results2, [f_sch, f_exa])
box_plot(df2, 'Vectorizer', 'Recall', ['Case', 'Columns', 'k'], angle_x=90, ax=axes, title='i2q Recall for Schema-based', order=vectorizers_order)
plt.savefig(f'../plots/blocking_real_exact_vec_sch.pdf', bbox_inches='tight')

fig, axes = plt.subplots(nrows=1, ncols=1)
df2 = filter_df(results2, [f_sch, f_exa])
box_plot(df2, 'Vectorizer', 'Precision', ['Case', 'Columns', 'k'], angle_x=90, ax=axes, title='i2q Precision for Schema-based', order=vectorizers_order)
plt.savefig(f'../plots/blocking_real_exact_vec_sch2.pdf', bbox_inches='tight')

fig, axes = plt.subplots(nrows=1, ncols=1)
df2 = filter_df(results2, [f_agn, f_exa])
box_plot(df2, 'Vectorizer', 'Recall', ['Case', 'Columns', 'k'], angle_x=90, ax=axes, title='i2q Recall for Schema-agnostic', order=vectorizers_order)
plt.savefig(f'../plots/blocking_real_exact_vec_agn.pdf', bbox_inches='tight')

fig, axes = plt.subplots(nrows=1, ncols=1)
df2 = filter_df(results2, [f_agn, f_exa])
box_plot(df2, 'Vectorizer', 'Precision', ['Case', 'Columns', 'k'], angle_x=90, ax=axes, title='i2q Precision for Schema-agnostic', order=vectorizers_order)
plt.savefig(f'../plots/blocking_real_exact_vec_agn2.pdf', bbox_inches='tight')
```

# Pearson

```python
df2 = filter_df(results2, [f_sch, f_exa])
df2 = df2[['Vectorizer', 'Case', 'Recall']]
#df2 = df2.set_index('Vectorizer')
df2['Total'] = df2.apply(lambda x: (x['Case'], x['Recall']), axis=1)
df2 = df2.groupby('Vectorizer')['Total'].apply(list)
df2 = df2.apply(lambda x: sorted(x, key=lambda x: x[0]))
df2 = df2.apply(lambda x: [y[1] for y in x])

no = df2.shape[0]
pearsonr = ones((no, no))

for i in range(no):
    for j in range(i+1,no):
        pearsonr[i,j] = pearsonr[j,i] = scipy.stats.pearsonr(df2[df2.index[i]], df2[df2.index[j]])[0]

ax = sns.heatmap(pearsonr)
ax.set_xticks(range(no), df2.index, rotation=90)
ax.set_yticks(range(no), df2.index, rotation=0)
```

```python
files = ['10K.csv', '50K.csv', '100K.csv', '200K.csv', '300K.csv', '1M.csv', '2M.csv']
markers = ['*', 'o', 'o', 'o', '*', '*', '^', '^', 'o', '*', '^', '*']

results = pd.read_csv('../logs/blocking_recall_euclidean_big.csv')
#results.Case = results.Case.apply(lambda x: files[x].split('.')[0])
results.Case = results.Case.apply(lambda x: files[x])
results['Case'] = results['Case'].apply(lambda x: x.split('.')[0])
results['Case'] = results['Case'].apply(lambda x: int(x[:-1])* (1000 if x[-1] == 'K' else 1000000) )
print(results.shape)
results = filter_df(results, [('k', 'eq', 10), f_i2q])
print(results.shape)
results
```
