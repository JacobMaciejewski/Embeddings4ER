```python
import json
import pandas as pd
import matplotlib.pyplot as plt

from utils import box_plot, filter_df, vectorizers_order, line_plot
```

```python
input_dir = '/home/alex/Desktop/repos/entity_matching/logs/'
```

# Real Data

```python
with open(input_dir+'vectorization_real.txt') as f:
    lines = []
    for line in f.readlines():
        line = json.loads(line)
        
        lines.append((line['dir'], line['vectorizer'], line['time'], line['file'], line['column']['name'],
                        line['memory']['process']['rss'], line['memory']['total']['used']))
    
df = pd.DataFrame(lines, columns=['Case', 'Vectorizer', 'Total Time', 'File',
                                        'Column', 'Memory Process', 'Memory Total'])

df['Memory Process'] = df['Memory Process'] / (1024*1024*1024)

fig, axes = plt.subplots(nrows=1, ncols=1)
box_plot(df, 'Vectorizer', 'Total Time', ['Case', 'File', 'Column'], 90, axes, "", order=vectorizers_order, yscale='log', ylabel='Total Time (sec) (log)')
#box_plot(df, 'Vectorizer', 'Total Time', ['Case', 'File', 'Column'], 90, axes, "", order=vectorizers_order, ylabel='Total Time (sec)')
plt.savefig(f'../plots/vectorization_real_time.pdf', bbox_inches='tight')

fig, axes = plt.subplots(nrows=1, ncols=1)
box_plot(df, 'Vectorizer', 'Memory Process', ['Case', 'File', 'Column'], 90, axes, "", order=vectorizers_order, ylabel='Memory Process (GBytes)')
plt.savefig(f'../plots/vectorization_real_memory.pdf', bbox_inches='tight')
```

```python
df.to_csv('vectorication_real.csv')
```

# Synthetic Data

```python
#files = ['10K.csv', '50K.csv', '100K.csv', '200K.csv', '300K.csv', '1M.csv', '2M.csv']
files = ['10K', '50K', '100K', '200K', '300K', '1M', '2M']
markers = ['*', 'o', 'o', 'o', '*', '*', '^', '^', 'o', '*', '^', '*']

with open(input_dir+'vectorization_big.txt') as f:
    lines = []
    for line in f.readlines():
        line = json.loads(line)
        lines.append((line['vectorizer'], line['time'], line['file'], line['memory']['process']['rss']))
    
df = pd.DataFrame(lines, columns=['Vectorizer', 'Total Time', 'File', 'Memory Process'])
df['File'] = df['File'].apply(lambda x: x.split('.')[0])
df['File'] = df['File'].apply(lambda x: int(x[:-1])* (1000 if x[-1] == 'K' else 1000000) )

df['Memory Process'] = df['Memory Process'] / (1024*1024*1024)


fig, axes = plt.subplots(nrows=1, ncols=1)
#line_plot(df, 'Vectorizer', 'Total Time', 'File', legend=False, ax=axes, order=files, markers=markers, yscale='log', ylabel='Total Time (log)')
line_plot(df, 'Vectorizer', 'Total Time', 'File', legend=True, ax=axes, markers=markers, yscale='log', ylabel='Total Time (sec) (log)', xlabel='#entities')
#line_plot(df, 'Vectorizer', 'Total Time', 'File', legend=False, ax=axes, markers=markers, ylabel='Total Time (sec)', xlabel='#entities')
plt.savefig(f'../plots/vectorization_synthetic_time.pdf', bbox_inches='tight')



fig, axes = plt.subplots(nrows=1, ncols=1)
#line_plot(df, 'Vectorizer', 'Memory Process', 'File', legend=True, ax=axes, order=files, markers=markers, ylabel='Memory Process (Bytes)')
line_plot(df, 'Vectorizer', 'Memory Process', 'File', legend=True, ax=axes, markers=markers, ylabel='Memory Process (GBytes)', xlabel='#entities')
plt.savefig(f'../plots/vectorization_synthetic_memory.pdf', bbox_inches='tight')
```
